## Setup inicial

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración visual
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Cargar dataset enriquecido
df = pd.read_csv("../data/processed/yellow_tripdata_enriquecido.csv", parse_dates=["tpep_pickup_datetime"])
```

## 1. Zonas de origen más frecuentes

```python
top_origen = df["PU_Zone"].value_counts().head(10)
sns.barplot(x=top_origen.values, y=top_origen.index, palette="viridis")
plt.title("Top 10 zonas de origen")
plt.xlabel("Cantidad de viajes")
plt.ylabel("Zona")
plt.tight_layout()
plt.show()
```

## 2. Zonas de destino más frecuentes

```python
top_destino = df["DO_Zone"].value_counts().head(10)
sns.barplot(x=top_destino.values, y=top_destino.index, palette="magma")
plt.title("Top 10 zonas de destino")
plt.xlabel("Cantidad de viajes")
plt.ylabel("Zona")
plt.tight_layout()
plt.show()
```

## 3. Heatmap de viajes entre zonas

```python
matriz = df.groupby(["PU_Zone", "DO_Zone"]).size().unstack(fill_value=0)
sns.heatmap(matriz, cmap="YlGnBu", linewidths=0.5)
plt.title("Heatmap de viajes entre zonas")
plt.xlabel("Zona destino")
plt.ylabel("Zona origen")
plt.tight_layout()
plt.show()
```

## . Serie temporal de viajes por día

```python
df["fecha"] = df["tpep_pickup_datetime"].dt.date
viajes_por_dia = df.groupby("fecha").size()
viajes_por_dia.plot(kind="line", marker="o", color="teal")
plt.title("Cantidad de viajes por día")
plt.xlabel("Fecha")
plt.ylabel("Número de viajes")
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 5. Boxplot de tarifas por zona de origen

```python
zonas_top = df["PU_Zone"].value_counts().head(10).index
df_filtrado = df[df["PU_Zone"].isin(zonas_top)]

sns.boxplot(x="PU_Zone", y="total_amount", data=df_filtrado, palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Distribución de tarifas por zona de origen")
plt.xlabel("Zona")
plt.ylabel("Total cobrado (USD)")
plt.tight_layout()
plt.show()
```
